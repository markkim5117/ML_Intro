# 3. Parametric and non-parametric models (5pt)

**(a) Describe the differences between a parametric and a non-parametric statistical learning approach.**

**(b) Advantages and disadvantages?**

(a) We can distinguish parametric from non-parametric models by its existence of parameters. Those two models are completely different. For parametric models, we have to assume exact form of functions such as degree, number of independent variables and so on. Linear regression and mulitple regression would be its examples. On the other hand, non-parametric models assume nothing but only depends on its sample(training) data. For instance, we can take a window and take an average of the points to get the value of f^.

(b) 
Advantage of parametric models: Since parametric models provide exact functions such as linear regression, it is more interpretable and we can find some implications if the model is successful.
Disadvantage of parametric models: Parametric model are restrictive. The model we choose will usually not match the true unknown form of f and a poor match results in a poor fit. So if the chosen model is too far from the true f, then our estimate will be poor.
Advantage of non-parametric models: We can avoid the function mis-match issue, suffered by the parametric methods. There is no need to assume a specific functional form of f.
Disadvantage of non-parametric models: A very large number of observations is required, because we cannot rely on the model with only few data. We can expect that the non-parametric would converge if there is a very large number of training sets. Also, there is no estimation of parameters so it is not  that interpretable.

# 10. Mini-batch gradient descent (20pt)

**In the Lab 2, we learned how to implement gradient descent and stochastic gradient descent. For this question, implement mini-batch gradient descent. Set the ````batch_size```` to 10.**

`````python
from numpy.linalg import inv
import statsmodels.api as smi

boston = pd.read_csv('data/Boston.csv') # load data

X_train = boston[["lstat"]].values # regression using all input columns
X_train = smi.add_constant(X_train)
y_train = boston['medv'].values
beta_direct = inv(X_train.T.dot(X_train)).dot(X_train.T).dot(y_train)
print(beta_direct)

beta = np.random.rand(2)*30
alpha = 0.001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent
n = float(len(X_train)) # Number of elements in X
batch_size = 10

######################## Write your answere here ########################
y_train = boston[['medv']].values
data = np.append(X_train, y_train, axis = 1)

def create_mini_batches():
    mini_batches = []
    n_mini_batches = int(data.shape[0] // batch_size) # 50
    np.random.shuffle(data)
    for i in range(n_mini_batches):
        mini_batch = data[i*batch_size:(i+1)*batch_size, :]
        mini_batches.append(mini_batch)
    return mini_batches

cost = []

# Performing gradient descent
for epoch in range(epochs):
    mini_batches = create_mini_batches()
    for mini_batch in mini_batches:
        X_mini = mini_batch[:, 0:2]
        y_mini = mini_batch[:, 2:]
        y_pred = np.dot(X_mini, beta.T).reshape((-1, 1))
        error = y_pred - y_mini
        dbeta = (1/batch_size) * np.dot(X_mini[:, 1:].T, error)
        beta = beta - (alpha * dbeta)
        temp = np.dot(error.T, error)
    J = (1/n)*temp[0][0]
    cost.append(J)

#########################################################################

print("Learned parameters through MBGD: {}".format(beta))
print("Directly learned parameters: {}".format(beta_direct))   
plt.plot(np.arange(epochs), cost)
plt.title("Learning curve")
plt.xlabel("Epoch")
plt.ylabel(r"J($\beta$)");
`````
